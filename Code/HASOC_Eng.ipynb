```python
d = {'luv':'love','wud':'would','lyk':'like','wateva':'whatever','ttyl':'talk to you later','kul':'cool','fyn':'fine','omg':'oh my god','fam':'family','bruh':'brother',
    'cud':'could','fud':'food','govt':'government'}
df['clean_text'] = df['clean_text'].apply(lambda x : ' '.join(d[word] if word in d else word for word in x.split()))
```

```python
#Now apply lemmatization to reduce the data to its base form
ps = nltk.PorterStemmer()
def stemming(list_of_words):
  stemmed_list = [ps.stem(word) for word in list_of_words]
  return stemmed_list
df['stemmed_words'] = df['tokenized_data'].apply(lambda x: stemming(x))
df.head()
# lemmatizer = WordNetLemmatizer()
# df['lemmatized_text'] = df['clean_text'].apply(lambda x : ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
```

```python
# df['stemmed_sentences'] = df['original_text']
#Now let's vectorize the data
tfidf_vectorizer = TFID(use_idf=True,max_features=5000)
train_vecs=tfidf_vectorizer.fit_transform(xtrain)
```

```python
import pickle
pickle.dump(tfidf_vectorizer,open('/content/drive/My Drive/feature.pkl',"wb")) 
```

```python
def create_vectorized_df(tfidf_vectorizer_vectors):
  ret_arr = []
  for i in range(tfidf_vectorizer_vectors.shape[0]):
    check = tfidf_vectorizer_vectors[i].T.todense()
    arr_d = (np.asarray(check)).flatten()
    ret_arr.append(pd.DataFrame(arr_d,index=tfidf_vectorizer.get_feature_names()))
  return ret_arr
```

```python
train_vectorized_array = create_vectorized_df(train_vecs)
train_vectorized_df = pd.concat(train_vectorized_array,axis=1)
train_vectorized_df.columns = [i for i in range(train_vectorized_df.shape[1])]
traindf = pd.DataFrame(data=train_vectorized_df).T
traindf
```

```python
test_vecs=tfidf_vectorizer.transform(xtest)
test_vectorized_array = create_vectorized_df(test_vecs)
test_vectorized_df = pd.concat(test_vectorized_array,axis=1)
test_vectorized_df.columns = [i for i in range(test_vectorized_df.shape[1])]
testdf = pd.DataFrame(data=test_vectorized_df).T
testdf
```

Task 1 RNN

```python
import tensorflow as tf
# The maximum number of words to be used. (most frequent)
MAX_NB_WORDS = 35000
# Max number of words in each message.
MAX_SEQUENCE_LENGTH = 50
# This is fixed.
EMBEDDING_DIM = 110
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=MAX_NB_WORDS, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)
tokenizer.fit_on_texts(ttdf['stemmed_sentences'].values)
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

```python
X = tokenizer.texts_to_sequences(df['stemmed_sentences'].values)
X = tf.keras.preprocessing.sequence.pad_sequences(X, maxlen=MAX_SEQUENCE_LENGTH)
print('Shape of data tensor:', X.shape)
```

```python
Y = pd.get_dummies(df['task1']).values
```

```python
from sklearn.model_selection import train_test_split
nxtrain, nxtest, nytrain, nytest = train_test_split(X,Y,test_size =0.1,stratify=Y,random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Embedding
```

```python
regressor = Sequential()
```

```python
regressor.add(Embedding(MAX_NB_WORDS, EMBEDDING_DIM, input_length=X.shape[1]))
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units = 40))
regressor.add(Dropout(0.2))
```

```python
regressor.add(Dense(2, activation='softmax'))
```

```python
regressor.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
epochs = 10
batch_size = 64
history = regressor.fit(nxtrain, nytrain, epochs=epochs, batch_size=batch_size,validation_split=0.1,callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, min_delta=0.0001)])
```

```python
accr = regressor.evaluate(nxtest,nytest)
accr[1]
```
