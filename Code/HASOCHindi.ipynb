Task 1 RNN

```python
import tensorflow as tf
# The maximum number of words to be used. (most frequent)
MAX_NB_WORDS = 35000
# Max number of words in each message.
MAX_SEQUENCE_LENGTH = 50
# This is fixed.
EMBEDDING_DIM = 110
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=MAX_NB_WORDS, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)
tokenizer.fit_on_texts(ttdf['clean_text'].values)
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

```python
X = tokenizer.texts_to_sequences(df['clean_text'].values)
X = tf.keras.preprocessing.sequence.pad_sequences(X, maxlen=MAX_SEQUENCE_LENGTH)
print('Shape of data tensor:', X.shape)
```

```python
Y = pd.get_dummies(df['task_1']).values
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Embedding
```

```python
from sklearn.model_selection import train_test_split
nxtrain, nxtest, nytrain, nytest = train_test_split(X,Y,test_size =0.1,stratify=Y,random_state=42)
```

```python
regressor = Sequential()
```

```python
regressor.add(Embedding(MAX_NB_WORDS, EMBEDDING_DIM, input_length=X.shape[1]))
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units = 40))
regressor.add(Dropout(0.2))
```

```python
regressor.add(Dense(2, activation='softmax'))
```

```python
regressor.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
epochs = 10
batch_size = 64
history = regressor.fit(nxtrain, nytrain, epochs=epochs, batch_size=batch_size,validation_split=0.1,callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, min_delta=0.0001)])
```

```python
accr = regressor.evaluate(nxtest,nytest)
accr[1]
```
